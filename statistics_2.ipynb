#### 1. In a survey of 100 people:
 - 40 like pizza (A)
 - 50 like burgers (B)
 - 20 like both (A ∩ B)
 - Find P(A), P(B), P(A ∩ B), and P(A | B).


```python
prob_of_a = 40 / 100
prob_of_b = 50 / 100
prob_of_a_and_b = 20 / 100
prob_of_a_given_b = prob_of_a_and_b / prob_of_b

print(f"P(A) = {prob_of_a}")
print(f"P(B) = {prob_of_b}")
print(f"P(A ∩ B) = {prob_of_a_and_b}")
print(f"P(A|B) = {prob_of_a_given_b}")
```
