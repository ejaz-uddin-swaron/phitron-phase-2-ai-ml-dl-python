### 1. Add New Column and Save File
You have a DataFrame with column Price.
 Create a new column Discounted_Price with (10% discounted price).
 Save the modified DataFrame as discounted_products.csv.

```python
import pandas as pd

df = pd.read_csv("Practice Day (1).csv")
df["Discounted_Price"] = df["Price"] - df["Price"] * 0.1
df.to_csv("discounted_products.csv")
```

### 2. Check Duplicates and Remove Them
Column: EmployeeID
 Tasks:
Count duplicate entries


Remove duplicates


Display cleaned DataFrame


```python
df = pd.read_csv("discounted_products.csv")
print(df["EmployeeID"].duplicated().sum())
df.drop_duplicates("EmployeeID", inplace=True)
df
```

### 3. Handle Missing Values
Columns: Age, Salary
 Tasks:
Count missing values


Fill missing Age with mean age


Drop rows where Salary is missing

```python
# len(df["Age"]) - df["Age"].nunique()
print(df["Age"].isnull().sum())
print(df["Salary"].isnull().sum())

df["Age"].fillna(df["Age"].mean(), inplace=True)
print(df)

df.dropna(subset=["Salary"], inplace=True)
print(df)
```

### 4. Apply Statistical Functions
Columns: Marks1, Marks2, Marks3
 Tasks:
Calculate mean, max for each subject


Compute total marks per student


Find standard deviation of total marks


```python
print(df["Marks1"].mean()) 
print(df["Marks2"].mean()) 
print(df["Marks3"].mean())

print(df["Marks1"].max()) 
print(df["Marks2"].max()) 
print(df["Marks3"].max()) 

df["Total Marks"] = df["Marks1"] + df["Marks2"] + df["Marks3"]
print(df["Total Marks"].std())
df
```

### 5. Histogram, Bar, Scatter, Pie
Create:   
 1. Histogram of Age
 2. Scatter plot (Age vs Salary)
 3. Scatter plot ( Price vs Discounted Price) 


```python
import matplotlib.pyplot as plt

plt.xlabel("Age")
plt.ylabel("Count")
plt.hist(df["Age"], color="red", edgecolor="black", label="Age of employees")
plt.legend()
plt.show()

plt.xlabel("Age")
plt.ylabel("Salary")
plt.scatter(df["Age"], df["Salary"])
plt.show()

plt.xlabel("Price")
plt.ylabel("Discounted Price")
plt.scatter(df["Price"], df["Discounted_Price"])
plt.show()
```
