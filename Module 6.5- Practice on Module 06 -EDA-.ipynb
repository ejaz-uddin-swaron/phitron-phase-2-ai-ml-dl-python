## Section 4 ‚Äî Identify Target Column and Feature Types

Separate the target and features into numeric and categorical groups.


```python
df["transaction_time"] = pd.to_datetime(df["transaction_time"], utc=True)

# TODO:
# 1. Identify your target column name
target_column = "is_fraud"

# 2. Create a list of numeric feature columns
numeric_columns = []

for col in df.columns:
    if df[col].dtype in ["int64", "float64", "datetime64[ns, UTC]"] and col != "is_fraud":
        numeric_columns.append(col)


# 3. Create a list of categorical feature columns
categorical_columns = []
    
for col in df.columns:
    if df[col].dtype == "object":
        categorical_columns.append(col)

# 4. Print all three
print("Target column: ", target_column)
print("Numeric column: ", numeric_columns)
print("Categorical column: ", categorical_columns)
```

### Student Notes and Observations

_Which columns did you treat as numeric? Which are categorical? Any columns that were hard to classify?_

_int64, float64 and datetime64[ns, UTC] were treated as numeric. object was treated as categorical columns. It was easy to detect columns._

## Section 10 ‚Äî Pairplot and Correlation Analysis

## Section 11 ‚Äî Categorical Features versus Target (Crosstab and Stacked Bar Plot)

Examine how target proportions change across categories.


```python
# TODO:
# For each categorical feature:
# 1. Create a crosstab between the feature and target with normalize="index"
for c in categorical_columns:
    ct = pd.crosstab(df[c], df[target_column], normalize="index")
    print(f"\nProportion of fraudulent within {c}")
    display(ct)

    ct.plot(kind="bar", stacked=True, figsize=(6,4))
    plt.title(f"fraudulent proportion by {c}")
    plt.ylabel("proportion")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

# 2. Plot a stacked bar chart of the proportions
# 3. Comment on how the target distribution changes across categories
```
