#### B1. Three scalers side by side
Heights = [150, 160, 170, 175, 180]
Weights = [58, 62, 65, 66, 190]
 Tasks:
  - a) Min-Max scale both to [0, 1]
  - b) Standardize the first three values of each only
  - c) Robust scale Weights with median and IQR
  - d) One line on which scaler handles the outlier best

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "heights":[150, 160, 170, 175, 180],
    "weights":[58, 62, 65, 66, 190]
})

h_small = df.min().iloc[0]
h_big = df.max().iloc[0]
w_small = df.min().iloc[1]
w_big = df.max().iloc[1]

df["height_scaled"] = (df["heights"]-h_small) / (h_big-h_small)
df["weight_scaled"] = (df["weights"]-w_small) / (w_big-w_small)

h_meu = df.loc[:,["heights","weights"]].mean().iloc[0]
h_sigma = df.loc[:,["heights","weights"]].std().iloc[0]
w_meu = df.loc[:,["heights","weights"]].mean().iloc[1]
w_sigma = df.loc[:,["heights","weights"]].std().iloc[1]

df["height_first_3_val_z_scaled"] = (df.loc[0:2,"heights"]-h_meu) / h_sigma
df["weight_first_3_val_z_scaled"] = (df.loc[0:2,"weights"]-w_meu) / w_sigma

w_median = df.loc[:,"weights"].median()
w_q1 = np.percentile(df.iloc[:,1],25)
w_q3 = np.percentile(df.iloc[:,1],75)

w_iqr = w_q3 - w_q1

df["weight_robust_scaled"] = (df.loc[:,"weights"] - w_median) / w_iqr
df
```

#### B2. One-hot by hand
Cities = [Dhaka, Chattogram, Dhaka, Rajshahi, Rajshahi]
Create three columns City_Dhaka, City_Chattogram, City_Rajshahi using 0 and 1.


```python
Cities = ["Dhaka", "Chattogram", "Dhaka", "Rajshahi", "Rajshahi"]

df = pd.DataFrame({
    "cities":Cities
})

df_city = pd.get_dummies(df.loc[:,"cities"], prefix="city", dtype=int)
df_encoded = pd.concat([df,df_city], axis=1)
df_encoded
```
