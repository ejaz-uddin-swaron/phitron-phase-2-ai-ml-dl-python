#### 1 . Create a 1D, 2D, and 3D NumPy array and print their dimensions and shapes.

```python
import numpy as np

arr_1d = np.random.randint(1,10,size=(3))
arr_2d = np.random.randint(1,10,size=(3,3))
arr_3d = np.random.randint(1,10,size=(3,3,3))

# print(arr_1d, end="\n\n")
# print(arr_2d, end="\n\n")
# print(arr_3d, end="\n\n")

print(arr_1d.ndim)
print(arr_2d.ndim)
print(arr_3d.ndim)

print(arr_1d.shape)
print(arr_2d.shape)
print(arr_3d.shape)

print(arr_1d.dtype)
print(arr_2d.dtype)
print(arr_3d.dtype)

print(arr_1d.size)
print(arr_2d.size)
print(arr_3d.size)
```

#### 6 . Given two arrays A and B, compute:
* (A + B) / 2
* A^2 + B^2


```python
arr_a = np.random.randint(1,10,(2,2))
arr_b = np.random.randint(1,10,(2,2))

print(arr_a, end="\n\n")
print(arr_b, end="\n\n")

ans1 = (arr_a + arr_b) / 2
ans2 = arr_a ** 2 + arr_b ** 2

print(ans1, end="\n\n")
print(ans2, end="\n\n")
```
