#### Part A – Central Tendency and Spread
1. Marks: 50, 60, 65, 70, 75, 80, 85, 90
Find mean, median, mode.
Replace 90 with 900 and recompute mean. Explain the effect of outliers.
2. Temperatures (°C): 29, 31, 33, 33, 32, 31, 30
Calculate variance and standard deviation.
Which measure is easier to interpret and why?


```python
import numpy as np
arr = np.array([50, 60, 65, 70, 75, 80, 85, 90])

mean = np.mean(arr)
median = np.median(arr)
dct = {}

for x in arr:
    dct[x] = dct.get(x,0) + 1

dct = sorted(dct.items(), key=lambda x:x[1], reverse=True) 

print(f"mean : {mean}")
print(f"median : {median}")
print(f"mode : {dct[0][0]}")

arr[arr == 90] = 900
print(arr)

print(f"Now mean : {np.mean(arr)}")

temperature = np.array([29, 31, 33, 33, 32, 31, 30])
variance = np.var(temperature)
standard_deviation = np.std(temperature)

print(f"Variance : {variance}")
print(f"Standarrd deviation : {standard_deviation}")

"""

Effect of the outlier (900):
The mean jumped from 71.875 to 173.125 — a huge change. That shows the mean is sensitive to outliers: a single extreme value drags the mean toward it. The median, however, stays 72.5 (unchanged) because it depends only on the middle rank(s). So, median is robust to outliers while mean is not.

Standard deviation is easier to interpret because it has the same units as the data (°C here). Variance is in squared units (°C²), which is less intuitive. So for practical interpretation, standard deviation is preferable.

"""
```

#### Part B – Percentiles, IQR and Z-Score
3. Cat weights (kg): 2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5
Find P25, P50, P75 and IQR. Mark possible outliers.
4. Given μ = 50, σ = 10
Find Z for x = 65 and interpret.
For Z = −2.5, find x.


```python
arr = np.array([2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5])
p25 = np.percentile(arr, 25)
p50 = np.percentile(arr, 50)
p75 = np.percentile(arr, 75)

print(f"p25 : {p25}")
print(f"p50 : {p50}")
print(f"p75 : {p75}")

Q1 = p25
Q2 = p50
Q3 = p75

IQR = Q3 - Q1

print(f"IQR : {p75-p25}")

print(f"Lower fence : {Q1 - (1.5 * IQR):.3f}")
print(f"Upper fence : {Q3 + (1.5 * IQR):.3f}")

"""

All data fall inside [2.325, 4.725].
Conclusion: No possible outliers by the 1.5·IQR rule.

"""

meu, sigma, x = 50, 10, 65

Z = (x - meu) / sigma
print(f"Z : {Z}")

# Now,

Z = -2.5

x = (Z * sigma) + meu

print(f"x : {x}")
```
